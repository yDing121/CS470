# Apriori

We will use the framework given by Professor Kai Shu's notes as a guideline for implementing the Apriori algorithm. We build a general algorithm that takes in a pandas dataframe and operates on one column.

```python
def parse_keywords(keywords: str):
    return set(keywords.split(";"))
```

```python
test_kws = df.iloc[0]["text_keywords"]
print(parse_keywords(test_kws))
print(type(parse_keywords(test_kws)))
# test_kws.split(";")
```

```python
def generate_1_itemsets(df: pd.DataFrame, target_col: str, min_support: int) -> set:
    counts = {}

    for kw_line in df[target_col]:
        kws = kw_line.split(";")

        for kw in kws:
            if kw in counts:
                counts[kw] += 1
            else:
                counts[kw] = 1

    ans = set()

    for k, v in counts.items():
        if v >= min_support:
            ans.add((k, v))

    return ans
```

```python
one_itemsets = generate_1_itemsets(df, "text_keywords", 500)
print(one_itemsets)
```

# Apriori Algorithm Implementation

We choose to use the F_k-1 x F_k scheme for candidate generation. The general process is as follows:



Steps:
1. Get $F_{k-1}$ and {F_k}
2. Merge them for all possible set pairwise set unions($C = \{a \cup b | a \in F_{k-1} , b \in F_k\}$)
3. Prune

```python
def parse_entry(entry) -> list:
    return entry.split(";")
```

```python
def apriori(df, column_name, min_support):
    # Drop entries with empty item sets
    entries = df[column_name].dropna()

    # Immutable sets for hashing (Optimization)
    parsed_entries = [frozenset(parse_entry(entry)) for entry in entries]

    # Get item counts
    counts = {}
    for entry in parsed_entries:
        for item in entry:
            counts[item] = counts.get(item, 0) + 1
    
    # F_1
    f_1 = {frozenset([item]): count for item, count in counts.items() if count >= min_support}
    
    f_k_minus_1 = f_1
    frequent_itemsets = [items_count_pair for items_count_pair in f_1.items()]
    k = 2
    
    while f_k_minus_1:
        candidates = set()
        f_k_minus_1_keys = list(f_k_minus_1.keys())
        
        # Candidate generation using F_k-1 x F_1
        for f_set in f_k_minus_1_keys:
            for item in f_1.keys():
                candidate = f_set | item
                if len(candidate) == k:
                    candidates.add(candidate)
        
        # Count candidates
        candidate_counts = {c: 0 for c in candidates}
        for entry in parsed_entries:
            for candidate in candidates:
                if candidate.issubset(entry):
                    candidate_counts[candidate] += 1
        
        # Prune
        f_k = {itemset: count for itemset, count in candidate_counts.items() if count >= min_support}
        
        if not f_k:
            break
        
        frequent_itemsets.extend([items_count_pair for items_count_pair in f_k.items()])
        f_k_minus_1 = f_k
        k += 1
    
    return frequent_itemsets
```

```python
measures = ["running_time", "num_itemsets", "comparison_count", "subset_check_count", "trie_insert_count"]
```

```python
# Set plot style
sns.set_style("whitegrid")
# Define marker styles
markers = {"Apriori_k1km1": "o", "Apriori_km1km1": "s", "Apriori_trie_improved": "D"}

# Generate a plot for each measure
for measure in measures:
    plt.figure(figsize=(12, 7))

    sns.lineplot(
        data=df, 
        x="min_support", 
        y=measure, 
        hue="algorithm", 
        style="algorithm",  # Different styles per algorithm
        markers=markers,    # Assign specific markers
        dashes=False,       # Use solid lines
        alpha=0.8,          # Reduce opacity to see overlapping points
        linewidth=2.5       # Thicker lines for better visibility
    )

    plt.title(f"Time Series Plot of {measure}")
    plt.xlabel("Min Support")
    plt.ylabel(measure)
    plt.xscale("log")  # Keeps spacing better if min_support varies widely
    plt.legend(title="Algorithm")
    plt.show()
```

```python
# Get running_time for each algorithm at each min_support
df_pivot = df.pivot(index="min_support", columns="algorithm", values="running_time").reset_index()

# Compute residuals relative to Apriori_km1km1
df_pivot["Apriori_k1km1_residual"] = df_pivot["Apriori_k1km1"] - df_pivot["Apriori_km1km1"]
df_pivot["Apriori_trie_improved_residual"] = df_pivot["Apriori_trie_improved"] - df_pivot["Apriori_km1km1"]
df_residuals = df_pivot.melt(id_vars="min_support", 
                             value_vars=["Apriori_k1km1_residual", "Apriori_trie_improved_residual"],
                             var_name="Algorithm", value_name="Residual Running Time")

# Plot residuals
plt.figure(figsize=(12, 6))
sns.line(data=df_residuals, x="min_support", y="Residual Running Time", hue="Algorithm", style="Algorithm")
plt.axhline(0, color="gray", linestyle="--")
plt.xscale("log")
plt.xlabel("Min Support")
plt.ylabel("Residual Running Time (compared to Apriori_km1km1) (secs)")
plt.title("Residual Plot of Running Time Compared to Apriori_km1km1")
plt.legend(title="Algorithm")
plt.show()
```
